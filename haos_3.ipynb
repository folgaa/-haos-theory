```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import phaseportrait
```

```python
a = 2

def vect(x0, y0):
  x = x0*(x0*(1-x0)-y0)
  y = y0*(x0-a)
  return (x, y)

def field_maker(x_arr, y_arr):
  for x in x_arr:
    for y in y_arr:
      xy = vect(x, y)
      plt.quiver(x, y, xy[0], xy[1], color = 'gray', width = 0.004)

def f(x,y):
  return [x*(x*(1-x)-y), y*(x - a)]

portrait = phaseportrait.PhasePortrait2D(f, [0, 2.5], xlabel="x", ylabel="y")
portrait.plot()

x = np.linspace(0, 2.5, 21)
y = np.linspace(0, 2.5, 21)
field_maker(x, y)
plt.show()
```
