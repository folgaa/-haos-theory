```python
#bifuraction diagram

import numpy as np
import matplotlib.pyplot as plt


def func(r, x):
  return r*((np.sin(x-1.2))**2)

def func_orbit(r, n):
  x = 0
  orbit = []
  for i in range(n+n//10):
    x = func(r, x)
    orbit += [x]
  return orbit[n//10:]

def bif_diagram(r, n):
  Y = []
  for R in r:
    y = func_orbit(R, n)
    Y += [y]
  plt.plot(r, Y, ls = '', marker = ',', color = 'k')
  plt.xlabel('r')
  plt.ylabel('X')
  plt.xlim(0,3)
  plt.ylim(0,)
  plt.show()

n = 1000
r = np.linspace(0, 3, 1001)
bif_diagram(r, n)
```

```python
#Lyapunov exponent

def func_der(r, x):
  return r*np.sin(2*x - 2.4)

def L_exp(r, n):
  L = 0
  x_i = np.array(func_orbit(r,n))
  f_i = np.abs(func_der(r, x_i))
  for f in f_i:
    if f != 0:
      L += np.log(f)
  L /= n
  return L

def L_plot(r, n):
  L = []
  for R in r:
    L += [L_exp(R, n)]
  plt.plot(r, L, color = 'k')
  plt.xlabel('r')
  plt.ylabel('X')
  plt.xlim(0,3)
  plt.ylim(-1,0.75)
  plt.grid()
  plt.show()

L_plot(r, n)
```

```python
#iteration diagram at r=0.5

def iter(r, x):
  plt.plot([x, x],[x, func(r, x)], color = 'g')
  return func(r, x)

r1 = 0.5
x0 = 0
x_pr = 0
n = 50

x = np.linspace(0, 0.5, 101)
y = func(r1, x)


plt.plot(x, y)
plt.plot(x, x)
for i in range(n):
  plt.plot([x_pr, x0],[func(r1, x_pr),func(r1, x_pr)], color = 'g')
  x_pr = x0
  x0 = iter(r1, x0)

plt.ylim(-0.1,0.7)
plt.show()
```
